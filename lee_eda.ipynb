### Q

```python
#Q로 T/V/M 나누기
eda_train['T'] = eda_train['Q1'] + eda_train['Q2'] + eda_train['Q3'] + eda_train['Q15']
eda_train['V'] = eda_train['Q4'] + eda_train['Q5'] + eda_train['Q7'] + eda_train['Q8'] + eda_train['Q10'] + eda_train['Q11'] + eda_train['Q12']  + eda_train['Q13'] + eda_train['Q17'] + eda_train['Q19'] + eda_train['Q20'] 
eda_train['M'] = eda_train['Q9'] + eda_train['Q16']
```

```python
#Q질문 이용하여 점수 계산(Q26개 설문의 답)
eda_train['Q_score'] = eda_train[Answers1].mean(axis = 1)
eda_train['Q_var'] = eda_train[Answers1].var(axis=1)
Ancoms = list(combinations(Answers1, 2))
temp_dict = {}
for a,b in Ancoms:
    temp_dict['%s_dv_%s'%(a,b)] = eda_train[a]/eda_train[b]
    
new_df = pd.DataFrame(temp_dict)
eda_train = pd.concat([eda_train, new_df], axis=1)
```

### TIPI

```python
#해당 column의 값을 뒤집는다 1~5 -> 5~1
flipping_columns = ["TIPI6", "TIPI2", "TIPI8", "TIPI4", "TIPI10"]
for flip in flipping_columns:
    eda_train[flip] = 6 - eda_train[flip]
```

```python
# TIPI test에 따라 항목 나누기  외향적 1~5   내향적 1~5 -> 5~1   내향적이 낮다 -> 외향적이 높다로 변환하였음 총 점수 2~10 가능 이를 /2 하여 1~5 값으로 통일
# 외향성, 친화성, 양심성, 정서적 안정, 개방성
eda_train['Ex'] = (eda_train['TIPI1']+eda_train['TIPI6'])/2
eda_train['Ag'] = (eda_train['TIPI7']+eda_train['TIPI2'])/2
eda_train['Con'] = (eda_train['TIPI3']+eda_train['TIPI8'])/2
eda_train['Es'] = (eda_train['TIPI9']+eda_train['TIPI4'])/2
eda_train['Op'] = (eda_train['TIPI5']+eda_train['TIPI10'])/2
```

```python
temp = eda_train[eda_train['gender'] == 3.0].copy()
temp['gender'] = 1.5
eda_train[eda_train['gender'] == 3.0] = temp.copy()
```

```python
#gender 3값 처리 필요 -> 1.5로 데이터 변환하였음
eda_train['Es_gender'] = eda_train['Es']*eda_train['gender']
eda_train['Con_gender'] = eda_train['Con']*eda_train['gender']
eda_train['Op_gender'] = eda_train['Op']*eda_train['gender']
```

### VLC

```python
#VLC질문을 이용하여 점수계산(단어의 정의 알고있는지 유무)
#VLC_score sum할 때 어려운 단어에 가중치 주기?
eda_train['VLC_score'] = eda_train[Answers3].sum(axis=1)       
eda_train['VLC_var'] = eda_train[Answers3].var(axis=1)
eda_train['VLC_mean'] = eda_train[Answers3].mean(axis = 1)    
```

```python
#VLC질문으로 스코어 냈으니 기존 VLC1~VLC16이 필요할까?
```

### elapse

```python
#지연시간
eda_train['delay'] = eda_train[['introelapse','testelapse','surveyelapse']].sum(axis=1)
eda_train['delay'] = eda_train['delay'] ** (1/10)
eda_train['delay_var'] = eda_train['delay'].var()
#sns.distplot(eda_train['delay'], rug = True)

#기존의 elapse 데이터 필요할까? OR 각각의 데이터를 분포하게 만들기?
eda_train.drop('introelapse', axis=1, inplace = True)
eda_train.drop('testelapse', axis=1, inplace = True)
eda_train.drop('surveyelapse', axis=1, inplace = True)
```

### Other Feature

```python
#age로 age_group object 생성
def age_grouping(age):
    age = (age//10.)*10
    if(age >= 70):
        age = '+70s'
    else:
        age = str(int(age))+'s'
    return age

eda_train['age_group'] = eda_train['age'].apply(age_grouping)
```
